```python
import pandas as pd
```

```python
#öncelikle manipülasyon için bir veri çerçevesi oluşturalım:
veri = {"isim":["x","y","z","w"],"yas":[38,24,19,51],"tutulan takım":["rm","juve","manu","liv"],"maas":[3000,4000,5000,2000]}
```

```python
df = pd.DataFrame(veri)
```

```python
df
```

```python
#VERİ ÇERÇEVESİNE BİR SÜTUN EKLEME
df["sehir"] = ["istanbul","izmir","ankara","malatya"]
```

```python
df
```

```python
#BİR SÜTUNUN İSMİNİ DEĞİŞTİRME
df = df.rename(columns = {"sehir":"il"})
```

```python
df
```

```python
#VERİ ÇERÇEVESİNE BİR SATIR EKLEME
#loc[] ile:
yeniSatir = {"isim":"q","yas":35,"tutulan takım":"mcity","maas":10000,"il":"bursa"}
```

```python
yeniSatir
```

```python
df.loc[len(df)] = yeniSatir #len(df), veri çerçevesinin uzunluğunu verir. bu sayede eklemeler son satıra yapılmış olur.
```

```python
df
```

```python
#concat() ile:
a = [{"isim":"h","yas":21,"tutulan takım":"mk","maas":0,"il":"kütahya"}] #oluşturduğumuz sözlüğü list içinde tutmalıyız.
yeniSatir = pd.DataFrame(a)
#DataFrame içine a yerine eklenecek satırı yazacaksak şöyle yazmalıyız: (köşeli parantez kullanmazsak hata veriyor)
#{"isim":["q"],"yas":[35],"tutulan takım":["mcity"],"maas":[10000],"il":["bursa"]}
```

```python
df = pd.concat([df, yeniSatir], ignore_index=True)
```

```python
df
```

```python
#VERİ ÇERÇEVESİNE BİRDEN FAZLA SATIR EKLEME
yeniSatirlar = [{"isim":"h","yas":21,"tutulan takım":"mk","maas":0,"il":"kütahya"},{"isim":"f","yas":75,"tutulan takım":"olum","maas":3500,"il":"kayseri"}]
```

```python
#loc[] ile:
for i in yeniSatirlar:
    df.loc[len(df)] = i
```

```python
df
```

```python
#concat() ile:
yeniSatirlar = [{"isim":"h","yas":21,"tutulan takım":"mk","maas":0,"il":"kütahya"},{"isim":"f","yas":75,"tutulan takım":"olum","maas":3500,"il":"kayseri"}]
ek = pd.DataFrame(yeniSatirlar)
df = pd.concat([df,ek])
```

```python
df
```

```python
#VERİ ÇERÇEVELERİNDE SATIR SİLME İŞLEMİ
#satır silme işlemi için drop() fonksiyonunu kullanırız. parametre olarak silinecek satırların indeks bilgilerini göndeririz.
son = df.drop([5,6,7])
```

```python
son
```

```python
#VERİ ÇERÇEVELERİNDE SÜTUN SİLME İŞLEMİ
silinecekSutun = ["tutulan takım"]
```

```python
son = son.drop(silinecekSutun,axis=1) #sütun sildiğimizi belirtmek için axis=1 yazdık
```
