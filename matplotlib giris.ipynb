# plt.subplot2grid() fonksiyonu
ızgaraların içeriine istenilen boyutlarda grafikler eklemek için kullanılan fonksiyondur. bu fonksiyon sırasıyla shape, location, rowspan, colspan parametrelerini alır.

- shape: satır ve sütun değerleri alır. birinci girdi satır sayısını ikinci girdi sütun sayısını gösterir.
- loc: 2 tam sayı alır. ızgara içerisinde nereye yerleştirileceğini belirler.
- rowspan: sağ tarafa doğru ne kadar genişleyeceğini belirtir.
- colspan: aşağı doğru ne kadar genişleyeceğini belirler.

```python
a1 = plt.subplot2grid((3,3),(0,0),colspan = 2) #3*3lük bir ızgarada 1. satır birinci sütunda iki sütun ilerler
a2 = plt.subplot2grid((3,3),(0,2),rowspan = 3)
a3 = plt.subplot2grid((3,3),(1,0),rowspan = 2,colspan=2)
x=np.arange(1,10)
a2.plot(x,x*x)
a2.set_title("square")
a1.plot(x,np.exp(x))
a1.set_title("exp")
a3.plot(x,np.log(x))
a3.set_title("log")
plt.tight_layout()
plt.show()
```

savefig() fonskiyonu

matplotlib kütüphanesinde grafikleri resim olarak kaydetmek, grafikleri göstermek kadar kolaydır. grafikleri jpeg.png gibi formatlarda kaydetmek için bu fonk. kullanılır. bu fonk. içerisine parametre olarak grafiği yolu ve ismini alır. bu parametre frame olarak verilir. frame dışında kullanılabiliecek diğer bazı parametreler aşağıda verilmektedir.

- dpi = None: görüntünün çözünürlüğü, örneğin 300dpi (inç başına düşen nokta sayısı)

png formatı internet kullanımları ve çizim grafikleri için daha uygundur. jpg formatı ise daha sıkılaştırılmış bir resim formatıdır. bu nedenle daha küçük boyutlarda görüntüleri kaydetmemizi sağlar. eğer daha küçük boyutlu resimlerle çalışılmak isteniliyorsa jpg formatı tercih edilebilir. svg formatı ise vektörel görüntülerde kullanılan ve diğer formatlara göre boyutu yüksek bir seçimdir.

# grafik çeşitleri
matplotlib içerisinde hemen hemen bilindik tüm grafik çeşidi çizdirilebilir. burada grafik çeşitlerine örnek olarak aşağıdaki çeşitler gösterilecektir:

1. Histogram: x eksenindeki verilerin dikdörtgen şekilde çizdirildiği grafik çeşididir. bu nedenle kesikli verilerle kullanılması gerekmektedir. eğer süreli veriler söz konusu ise sürekli veriler bins denilen çerçevelere ayrılır. hist() fonk ile çizdirilebilir. hist() fonk.u n, bins, patches gibi parametreler alır. yani histogram ile gruplu verilerin grafiği çizdirilebilir. örneğin kadın ve erkeklere ilişkin yaşlar histogram grafiği olarak aşağıdaki gibi çizdirilir.

```python
kadin_yaslar = np.random.randint(low=1,high=75,size=100)
plt.hist(kadin_yaslar,label="kadın yaşları")
plt.legend()
```

```python
erkek_yaslar = np.random.randint(low=1,high=75,size=100)
plt.hist(kadin_yaslar,label="kadın yaşları",alpha=0.5)
plt.hist(erkek_yaslar,label="erkek yaşları",alpha=0.5)
plt.legend()
```

```python
plt.hist([erkek_yaslar,kadin_yaslar],label=["erkek","kadın"])
plt.legend()
```

aynı şekilde histogram grafikleri olasılık yoğunluk fonksiyonlarının çizdirilmesinde de kullanılabilir. örneğin normal dağılıma uyan bir veri dizisinin grafiği aşağıdaki gibi çizdirilebilir.

```python
np.random.seed(12345)
x_normal = np.random.normal(size=100) #normal dağılıma uyan 100 veri
plt.hist(x_normal, density=True, bins=30) #density = false sayıları alacaktı, biz oranları aldık. bins=30, 30 luk kısımlar halinde al demek.
plt.ylabel("olasılıklar")
plt.xlabel("veriler")
```

```python
np.random.seed(12345)
x_normal = np.random.normal(size=100000) 
plt.hist(x_normal, density=True, bins=30)
plt.ylabel("olasılıklar")
plt.xlabel("veriler")
```

```python
x1_koordinatları = np.random.randint(10,size=20)
y1_koordinatları = np.random.randint(5,size=20)

x2_koordinatları = np.random.randint(10,size=20)
y2_koordinatları = np.random.randint(5,size=20)

plt.bar(x1_koordinatları,y1_koordinatları,label="grup 1")
```
